```python
import pandas as pd
import cufflinks as cf
import plotly.offline

# Load data
df = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv')

# Enable cufflinks offline mode
cf.go_offline()
cf.set_config_file(world_readable=True, theme='pearl')

# Plot
df[['sepal_length', 'sepal_width']].iplot(kind='scatter', mode='markers', title='Sepal Dimensions')
```

```python
df.iplot()
```

```python
df.iplot(x = 'sepal_length' , y = 'petal_length')
```

```python
df.iplot(x = 'sepal_length' , y = 'petal_length' , mode = 'markers')
```

```python
df.iplot(x = 'sepal_length' , y = 'petal_length' , mode = 'markers' , kind = 'scatter')
```

```python
df.iplot(mode = 'markers' , kind = 'scatter')
```

```python
df.iplot(kind = 'bubble' , x = 'sepal_length' , y = 'petal_length' , size = 'sepal_width' )
```
